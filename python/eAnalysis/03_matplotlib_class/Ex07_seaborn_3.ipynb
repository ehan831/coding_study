추가 그래프 확인

```python
import seaborn as sns                
import matplotlib.pyplot as plt  

tips = sns.load_dataset('tips')
tips.head()
```

(1) 시간대별로 지불금액의 평균을 나타내는 바이올린 그래프에 성별을 추가한다면

```python
plt.subplot
sns.violinplot(x='time', y='total_bill', data=tips, hue='sex')
```

(2) 지불금액과 팁에 대한 산점도 그래프에 성별을 추가한다면 

```python
plt.subplot
# sns.regplot(x='total_bill', y='tip', data=tips) # hue 가 없음, ㅠ
# ax = sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', fit_reg=False, markers=['o', 'x'])
ax = sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', fit_reg=False, scatter_kws={'s':tips['size'] * 20})
```

FacetGrid()를 이용한 그룹별 그래프

```python
# 시간대별로 지불 금액을 표현한 그래프
facet = sns.FacetGrid(tips, col='time')
facet.map(sns.distplot, 'total_bill')
```

```python
# 요일별로 지불 금액을 표현한 그래프에 성별 구분 추가
facet = sns.FacetGrid(tips, col='day', hue='sex')
# facet.map(sns.regplot, 'total_bill', 'tip')  # 아래처럼 써도 됨. sns는 plt 기반임
facet.map(plt.scatter, 'total_bill', 'tip')
facet.add_legend()
```

```python
# tips.columns  # Index(['total_bill', 'tip', 'sex', 'smoker', 'day', 'time', 'size'], dtype='object')
# 시간대별 흡연여부별 지불 금액을 표현한 그래프에 성별 구분 추가
facet = sns.FacetGrid(tips, col='time', row='smoker', hue='sex')
facet.map(sns.distplot, 'total_bill')
facet.add_legend()
```

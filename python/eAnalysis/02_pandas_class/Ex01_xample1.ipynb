```python
import pandas as pd

data = pd.read_csv('data/president_heights.csv')

data.head(10)
```

```python
data.info()
```

```python
# 기본 통계 ( 평균/표준편차/최소값/최대값/중앙값 )
print(data.mean())
print('=' * 20)
print(data.std())
print('=' * 20)
print(data.max())
print('=' * 20)
print(data.min())
print('=' * 20)
print(data.median())
```
